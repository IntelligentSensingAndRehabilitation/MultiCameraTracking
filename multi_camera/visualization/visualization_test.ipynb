```python
from IPython.display import HTML, clear_output
from etils import epath
import jinja2
```

```python
html_path = './index.html'
js_url = './js/viewer.js'
#js_url = 'http://127.0.0.1:8888/lab/tree/multi_camera/visualization/js/viewer.js'

#base_url = 'https://cdn.jsdelivr.net/gh/google/brax'
#js_url = f'{base_url}@v0.9.0/brax/visualizer/js/viewer.js'

# read the contents of html_path into variable html_template
with open(html_path, 'r') as f:
    html_template = f.read()

template = jinja2.Template(html_template)
html = template.render(js_url=js_url, colab=False,height=480, system_json={'meshes': [], 'geoms': [], 'states': []})

# display(html)

display(HTML(html))
```
